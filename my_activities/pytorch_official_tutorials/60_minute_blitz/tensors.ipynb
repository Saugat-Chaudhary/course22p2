Tensors
=======

Tensors are a specialized data structure that are very similar to arrays
and matrices. In PyTorch, we use tensors to encode the inputs and
outputs of a model, as well as the model's parameters.

Tensors are similar to NumPy's ndarrays, except that tensors can run on
GPUs or other specialized hardware to accelerate computing. If you're
familiar with ndarrays, you'll be right at home with the Tensor API. If
not, follow along in this quick API walkthrough.


```python
import torch
import numpy as np
```

## Tensors and their rankings

### Rank 0 tensor: Scalar
Scalar are just a number and are represented as rank 0 tensors.

```python
# rank 0 tensors: Scalar
# just a single number
scalar = torch.tensor(2)
print(f'Rank 0 (Scalar) ndim: {scalar.ndim}')
print(f'tensor: {scalar}')
```

### Rank 1 tensor: Vector
Vector are just 1D array of numbers.

```python
# rank 1 tensor: Vector (1D array)
vector = torch.tensor([1,2])
print(f'Rank 1 (Vector) ndim: {vector.ndim}')
print(f'tensor: {vector}')
```

### Rank 2 tensor: Matrix
A 2D grid (Rows, Columns)

```python
# rank 2 tensor: Matrix (2D array)
matrix = torch.tensor([[0,0],[0,1]])
print(f'Rank 2 (Matrix) ndim: {matrix.ndim}')
print(f'tensor: {matrix}')
```

## Rank 3 tensor: 3D Tensor
Often used for a single RGB image (Channels, Height, Width)

```python
# rank 3 tensor: 3D Tensor
tensor_3d = torch.zeros((3, 224, 224)) #(Channels, Height, Width)

# This prints the DATA (the zeros)
print(f'tensor: {tensor_3d}')

# This prints the RANK (should be 3)
print(f'Rank (ndim): {tensor_3d.ndim}')

# This prints the SHAPE (should be torch.Size([3, 224, 224]))
print(f'Shape: {tensor_3d.shape}')
```

### Rank 4 tensor: 4D Tensor
Commonly used for a batch of images (Batch, Channels, Height, Width)

```python
# rank 4 tensor: 4D Tensor
tensor_4d = torch.randn((32,3,224,224))
print(f'Rank 4 Tensor ndim: {tensor_4d.ndim}')
print(f'tensor: {tensor_4d}')
```

###_Note: why we used .zeros or .ones or .rands to create 3d or 4d tensor?_

| Function | What it does | Common Use Case |
|---------|--------------|-----------------|
| torch.zeros()| Fills with $0$ |Creating masks, padding, or empty image placeholders. |
| torch.ones() | Fills with $1$ | Initializing multipliers or "white" background masks. |
| torch.randn() | Fills with random numbers between -1 and 1 | Initializing model weights or simulating real-world data. |
| torch.rand() | Fills with random numbers between 0 and 1 | Initializing model weights or simulating real-world data. |

The simple answer is, we used .zeros(), .ones() or .randn() because it isn't possible for us to write a whole 3d or 4d tensor by typing as they are very big.

Tensor Initialization
=====================

Tensors can be initialized in various ways. Take a look at the following
examples:

**Directly from data**

Tensors can be created directly from data. The data type is
automatically inferred.


```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
x_data
```

**From a NumPy array**

Tensors can be created from NumPy arrays (and vice versa - see
`bridge-to-np-label`{.interpreted-text role="ref"}).


```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
x_np
```

**From another tensor:**

The new tensor retains the properties (shape, datatype) of the argument
tensor, unless explicitly overridden.

- `x_ones`: We are saying we want tensors in the form of one but maintain the shape and datatype of x_data tensor.
- `x_rand`: We are saying we want random tensors with float as datatype but still maintain the shape of x_data tensor.

```python
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

**With random or constant values:**

`shape` is a tuple of tensor dimensions. In the functions below, it
determines the dimensionality of the output tensor.

- shape = (2,3,): we are saying, we need rank 2 tensor (matrix) with 2 rows and 3 columns. The trailing comma doesn't have any special meaning, it is same as saying (2,3).

_Note: Major catch, if we didn't specify the datatype for tensor then by default the pytorch will create tensor of float32 type as neural networks need precision._


```python
shape = (2, 3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

Tensor Attributes
=================

Tensor attributes describe their shape, datatype, and the device on
which they are stored.

Tensors are usually stored in two types of devices:
1. CPU
2. GPU (cuda)
And if one tensor is in cpu and another is in gpu then any operation between those tensors will throw error so, it is important to check using method `tensor.device`.

```python
tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

**Multiplying tensors**
- **Element wise multiplication**

  ELement wise matrix multiplication requires both matrices to be of the same shape i.e. same number of rows and column.

  Operation: Each element in the result is A[i][j] * B[i][j].

  Matrix A:
  $[[1, 2, 3],
    [4, 5, 6]]$

  Matrix B:
  $[[ 7,  8,  9],
    [10, 11, 12]]$

  Hadamard Product:
  $[[ 7, 16, 27],
    [40, 55, 72]]$



```python
# This computes the element-wise product
print(f"tensor.mul(tensor) \n {tensor.mul(tensor)} \n")
# Alternative syntax:
print(f"tensor * tensor \n {tensor * tensor}")
```

**Matrix Multiplication:**

This computes the matrix multiplication between two tensors:

Matrix multiplication is an operation where two matrices are combined to form a new matrix. The general rule is that if you have matrix
A of size m × n and matrix B of size n × p, their product C=AB will be of size m×p. Each element $^cij$ in the result is calculated by taking the dot product of the i-th row of A and the j-th column of B.

Example: 2×3 and 3×2 Matrices

Let A =$ [[1, 0, 2, ],[−1, 3, 1]],

B = [[3, 1, 2, [1, 1, 0]]

The resulting matrix C will have size 2×2.

Step 1: Compute each element

$^c11$=1⋅3+0⋅2+2⋅1=3+0+2=5

$^c12$=1⋅1+0⋅1+2⋅0=1+0+0=1$

$^c21$=−1⋅3+3⋅2+1⋅1=−3+6+1=4

$^c22$=−1⋅1+3⋅1+1⋅0=−1+3+0=2

Step 2: Result

C=AB=$ [ [5, 1],[ 4 ,2] ]$

```python
print(f"Tensor.T: {tensor.T}")
print(f"tensor.matmul(tensor.T) \n {tensor.matmul(tensor.T)} \n")
# Alternative syntax:
print(f"tensor @ tensor.T \n {tensor @ tensor.T}")
```

**In-place operations** Operations that have a `_` suffix are in-place.
For example: `x.copy_(y)`, `x.t_()`, will change `x`.

In PyTorch, you can usually spot an in-place operation by the underscore (_) at the end of the function name.

- Standard Operation (Not in-place): y = x.add(5) — This leaves x alone and creates a brand new tensor y.

- In-Place Operation: x.add_(5) — This modifies the values inside x itself.


```python
x = torch.tensor([1.0, 2.0])

# 1. Standard (Out-of-place)
y = x + 1  # Or x.add(1)
print(f"Standard - x: {x}, y: {y}")
# x is still [1, 2], y is [2, 3]

# 2. In-place
x.add_(1)
print(f"In-place - x: {x}")
# x has now changed to [2, 3]

print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

Bridge with NumPy {#bridge-to-np-label}
=================

Tensors on the CPU and NumPy arrays can share their underlying memory
locations, and changing one will change the other.


Tensor to NumPy array
=====================


```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

A change in the tensor reflects in the NumPy array.


```python
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

NumPy array to Tensor
=====================


```python
n = np.ones(5)
t = torch.from_numpy(n)
```

Changes in the NumPy array reflects in the tensor.


```python
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
